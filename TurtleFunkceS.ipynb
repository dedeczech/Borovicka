```python
def dashed_line(delka, dash_length=10, gap_length=5):
    """Nakreslí čárkovanou čáru."""
    while delka > 0:
        pendown()
        forward(min(dash_length, delka))
        delka -= dash_length

        if delka <= 0:
            break

        penup()
        forward(min(gap_length, delka))
        delka -= gap_length

    pendown()   # po skončení vrátíme pero dolů
```

### Kružnice, hypnotická mandala

```python
def muj_kruh(step):
    """Nakreslí 36-úhelník, který vypadá jako kruh."""
    for _ in range(36):
        forward(step)
        left(10)

def mandala(step=15, colors=None):
    """
    Nakreslí mandalu z 36-úhelníků.

    step   - délka strany každého 36-úhelníku
    colors - seznam barev; pokud None, použije defaultní barevný tuple
    """
    if colors is None:
        colors = [(255,0,0), (255,127,0), (255,255,0),
                  (0,255,0), (0,255,255), (255,0,255)]

    for i in range(36):
        color(colors[i % len(colors)])
        muj_kruh(step)
        left(10)  # pootočení mandaly

# -------------------------------
initializeTurtle(initial_speed=13, initial_window_size=(500,300))
speed(10)

# Příklad použití
mandala(step=15)
```

```python
from math import pi

def circle(r, step=5):
    """
    Nakreslí kružnici aproximovanou malými úseky.
    r     - poloměr
    step  - délka jednoho úseku (čím menší, tím hladší)
    """
    # počet kroků (zaokrouhleno)
    n = int(2 * pi * r/ step)

    # úhel otočení v každém kroku
    turn_angle = 360 / n

    # přesun na začátek kružnice
    penup()
    forward(r)
    left(90)
    pendown()

    # vlastní kreslení bez trigonometrie
    for _ in range(n):
        forward(step)
        left(turn_angle)

    # návrat do středu
    penup()
    home()
    #right(90)
    #backward(r)
    pendown()
```

### Flower
arc(...) kreslí oblouk rozdělený na malé rovné úseky

petal(...) jen nakreslí dva stejné oblouky otočené proti sobě

flower(...) jen otáčí želvu o 360/petals a kreslí znovu

```python
def arc(radius, angle=60, steps=30):
    """Nakreslí oblouk daného poloměru a úhlu."""
    step_length = 2 * 3.14 * radius * (angle/360) / steps
    step_angle = angle / steps
    for _ in range(steps):
        forward(step_length)
        left(step_angle)

def petal(radius):
    """Jeden lístek – dva oblouky zpět k sobě."""
    arc(radius, angle=60)
    left(180-60)
    arc(radius, angle=60)
    left(180-60)

def flower(petals, petal_size, center_x=200, center_y=200):
    """Nakreslí květinu tvořenou počtem petals."""
    penup()
    goto(center_x, center_y)
    pendown()

    for _ in range(petals):
        petal(petal_size)
        left(360/petals)
```

```python
initializeTurtle(initial_window_size=(500,300))
speed(13)

flower(8, 60, center_x=300, center_y=200)
```

### Turtle + Gauss = abstraktní pointilismus
Výsledná vizualizace představuje stochasticky generovaný bodový rozptyl, v němž horizontální pozice vzorků sleduje aproximaci Gaussovy distribuce a barevná modalita osciluje v omezeném disjunktním spektru

```python
!pip install ColabTurtle
from ColabTurtle.Turtle import *
import math, random

initializeTurtle()
speed(13)
hideturtle()

canvas_w = 400
canvas_h = 400

colors = [
    (255,0,0),(255,128,0),(255,255,0),(128,255,0),
    (0,255,0),(0,255,128),(0,255,255),(0,128,255),
    (0,0,255),(128,0,255),(255,0,255),(255,0,128)
]

def set_color(idx):
    pencolor(colors[idx % len(colors)])

# --- SPIROGRAF FUNKCE (rychlá) ---
def spiro(R,r,d,cx,cy):
    penup()
    for t in range(0,360*3,6):  # rychlý krok → instantní
        theta = math.radians(t)
        x = (R - r) * math.cos(theta) + d * math.cos(((R - r)/r)*theta)
        y = (R - r) * math.sin(theta) - d * math.sin(((R - r)/r)*theta)
        goto(max(0,min(canvas_w,cx+x)), max(0,min(canvas_h,cy+y)))
        if t==0: pendown()
        pensize(2)
        if t % 50 == 0: set_color(t//50)

# --- FRAKTÁLNÍ STROM FUNKCE ---
def fraktal(delka,uhel,hloubka):
    if hloubka==0: return
    forward(delka)
    left(uhel)
    fraktal(delka*0.7,uhel,hloubka-1)
    right(2*uhel)
    fraktal(delka*0.7,uhel,hloubka-1)
    left(uhel)
    backward(delka)

# --- INSTANT BANGER ---
# Spirografy
for i in range(15):  # více vrstev → full-page
    R=random.randint(50,130)
    r=random.randint(10,60)
    d=random.randint(10,80)
    cx=random.randint(R+d,canvas_w-(R+d))
    cy=random.randint(R+d,canvas_h-(R+d))
    spiro(R,r,d,cx,cy)

# Fraktální stromy
for i in range(10):
    penup()
    goto(random.randint(50,canvas_w-50), random.randint(50,canvas_h-50))
    setheading(random.randint(0,360))
    pendown()
    set_color(i*3)
    fraktal(random.randint(20,60),random.randint(15,35),3)

hideturtle()
```
