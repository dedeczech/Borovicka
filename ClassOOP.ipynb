```python
class zarizeni:
    def __init__(self, nazev, vyrobce, zapnuto=False):
        self.nazev = nazev
        self.vyrobce = vyrobce
        self.zapnuto = zapnuto

    def zapni(self):
        if not self.zapnuto:
            self.zapnuto = True
            print(f"{self.nazev} bylo zapnuto.")
        else:
            print(f"{self.nazev} je již zapnuto.")

    def vypni(self):
        if self.zapnuto:
            self.zapnuto = False
            print(f"{self.nazev} bylo vypnuto.")
        else:
            print(f"{self.nazev} je již vypnuto.")

class Telefon(zarizeni):
    def __init__(self, nazev, vyrobce, operacni_system, zapnuto=False):
        super().__init__(nazev, vyrobce, zapnuto)
        self.operacni_system = operacni_system

    def info(self):
        stav = "zapnuto" if self.zapnuto else "vypnuto"
        print(f"Telefon: {self.nazev}, Výrobce: {self.vyrobce}, OS: {self.operacni_system}, Stav: {stav}")


class Notebook(zarizeni):
    def __init__(self, nazev, vyrobce, ram, zapnuto=False):
        super().__init__(nazev, vyrobce, zapnuto)
        self.ram = ram

    def info(self):
        stav = "zapnuto" if self.zapnuto else "vypnuto"
        print(f"Notebook: {self.nazev}, Výrobce: {self.vyrobce}, RAM: {self.ram} GB, Stav: {stav}")


class Televize(zarizeni):
    def __init__(self, nazev, vyrobce, uhlopricka, zapnuto=False):
        super().__init__(nazev, vyrobce, zapnuto)
        self.uhlopricka = uhlopricka # Corrected attribute name from uhloprickam to uhlopricka

    def zmen_kanal(self, kanal):
        if self.zapnuto:
            print(f"{self.nazev} přepíná na kanál {kanal}.")
        else:
            print(f"{self.nazev} je vypnutá, nelze přepnout kanál.")

    def info(self):
        stav = "zapnuto" if self.zapnuto else "vypnuto"
        print(f"Televize: {self.nazev}, Výrobce: {self.vyrobce}, Úhlopříčka: {self.uhlopricka} palců, Stav: {stav}")

t = Telefon("iPhone 16", "Apple", "iOS")
n = Notebook("ThinkPad", "Lenovo", 16)
tv = Televize("Bravia", "Sony", 24)

t.info()
t.zapni()

n.info()
n.zapni()

tv.info()
tv.zapni()
tv.zmen_kanal(5)
```
