```python
moje_list = ["jablko", "banán", "třešeň"]
print("Původní list:", moje_list)

moje_list.append("hruška")
print("Po přidání položky:", moje_list)

prvni = moje_list.pop(0)
print(f"Odstraněn první prvek: {prvni}")
print("List po odstranění:", moje_list)

print("Délka listu:", len(moje_list))

print("Aktuální list:", moje_list)
```
